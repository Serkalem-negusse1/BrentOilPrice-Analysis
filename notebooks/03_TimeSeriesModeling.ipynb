# Time Series Modeling

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load processed data
df = pd.read_csv("../results/processed_data.csv", parse_dates=['Date'])
df.set_index('Date', inplace=True)
```

```python
# Load forecast results
forecast_df = pd.read_csv("../results/forecasting_results.csv", parse_dates=['Date'])
```

```python
# Plot actual vs predicted prices

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Price'], label="Actual Price", color="blue")
plt.plot(forecast_df['Date'], forecast_df['ARIMA_Forecast'], label="ARIMA Forecast", linestyle="dashed", color="red")
plt.plot(forecast_df['Date'], forecast_df['GARCH_Forecast'], label="GARCH Forecast", linestyle="dashed", color="green")

plt.xlabel("Year")
plt.ylabel("Price (USD)")
plt.title("Brent Oil Price Forecast (Next 30 Days)")
plt.legend()
plt.show()
```
