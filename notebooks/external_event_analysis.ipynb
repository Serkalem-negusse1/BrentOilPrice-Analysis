# Event Impact Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load processed data
df = pd.read_csv("../data/processed_data.csv", parse_dates=['Date'])
df.set_index('Date', inplace=True)
```

```python
# Load external events
events_df = pd.read_csv("../results/event_impact_analysis.csv")
```

```python
# Convert event dates to datetime
events_df['Date'] = pd.to_datetime(events_df['Date'])
```

```python
# Plot oil price with external events

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Price'], label="Brent Oil Price", color="blue")

# Highlight external events
for _, row in events_df.iterrows():
    plt.axvline(row["Date"], color='red', linestyle="--", label=row["Event"])

plt.xlabel("Year")
plt.ylabel("Price (USD)")
plt.title("Brent Oil Price & External Events")
plt.legend()
plt.show()

# Display event impact analysis results
print(events_df)
```
